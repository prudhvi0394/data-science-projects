```python
#data ingestion

df_equity  = pd.read_csv("equity_value_data.csv",index_col=0)
df_features= pd.read_csv("features_data.csv",index_col=0)
```

```python
#reset index
df_equity=df_equity.reset_index()

#converting timestamp into datetime
df_equity['timestamp'] = pd.to_datetime(df_equity['timestamp'])
```

```python
"""calculating the difference between records in the equity table will tell us 
the amount of days the users equity fell below 10 which will tell us which users to mark as churned based
on the requirement"""

df_equity.sort_values(['user_id', 'timestamp'], inplace=True)

df_equity['gap_between_trx'] = df_equity.groupby(['user_id'])['timestamp'].transform(lambda x: x.diff()) 

#converting it into float from the timedelta format
df_equity['gap_between_trx']=df_equity['gap_between_trx'].dt.days
```

```python
#total users in the dataset
len(set(df_equity.user_id))
```

```python
#filtering out users who have arrived within 28 days of the last date in the dataset 
# since they aren't in the system for the observable window criteria let's separate them for accurate churn representation

len(set(df_equity[df_equity.timestamp<(df_equity.timestamp.max()-timedelta(28))].user_id))
```

```python
df_equity_filtered=df_equity[df_equity.timestamp<(df_equity.timestamp.max()-timedelta(28))]
```

```python
#finding the max of number of days between transaction at each user level
churned_users=pd.DataFrame(df_equity_filtered.groupby(['user_id'])['gap_between_trx'].max())
```

```python
#there are 5 users with Nan as the max difference in days
churned_users.isnull().sum()
```

```python
#finding out what's the deal with these people
a=churned_users[churned_users.gap_between_trx.isnull()].index
df_equity_filtered[df_equity_filtered.user_id.isin(a)]
```

## we can see that these users have only one record in the system and on taking a closer look at the data we see that they have been there in the system longer than 28 days before the end date of the window in which we are measuring churn for which means that these users have churned according to our definition

```python
#getting the list of churned users
churned_users[(churned_users.gap_between_trx>=28) | (pd.isnull(churned_users.gap_between_trx))]
```

```python
#percentage of churn
number_of_churned_users=len(churned_users[(churned_users.gap_between_trx>=28) | (pd.isnull(churned_users.gap_between_trx))])
total_users=len(set(df_equity.user_id))
percentage=(number_of_churned_users/total_users)*100
```

```python
#assigning flags to the churned users

churned_user_ids=churned_users[(churned_users.gap_between_trx>=28) | (pd.isnull(churned_users.gap_between_trx))].index
```

```python
#assigning 1 to churned users
df_equity.loc[df_equity.user_id.isin(churned_user_ids),'flag']=1
```

```python
#assigning 0 values to the unchurned users
df_equity.flag.fillna(0,inplace=True)
```

```python
assign=pd.DataFrame(df_equity.groupby(['user_id']).flag.mean())
```

```python
assign=assign.reset_index()
```

```python
#verifying authenticity
set(df_equity.user_id)-set(assign.user_id)
```

```python
#split into test train
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
#using different classifiers

modelLR = LogisticRegression()
modelLR.fit(x_train, y_train)

modelRF = RandomForestClassifier()
modelRF.fit(x_train, y_train)

modelGB = GradientBoostingClassifier()
modelGB.fit(x_train, y_train)
```

```python
modelLR = LogisticRegression(class_weight="balanced")
modelLR.fit(x_train, y_train)
```

```python
#classifier without weights
predictLR = modelLR.predict(x_test)
print('Logistic Regression: \n', classification_report(predictLR, y_test))
print('Logistic Regression Accuracy: ', accuracy_score(predictLR, y_test))

predictRF = modelRF.predict(x_test)
print('Random Forest Classifier : \n', classification_report(predictRF, y_test))
print('Random Forest Classifier Accuracy: ', accuracy_score(predictRF, y_test))

predictGB = modelGB.predict(x_test)
print('Gradient Boost Classifier : \n', classification_report(predictGB, y_test))
print('Gradient Boost Classifier Accuracy: ', accuracy_score(predictGB, y_test))
```

```python
#classifier with weights
predictLR = modelLR.predict(x_test)
print('Logistic Regression: \n', classification_report(predictLR, y_test))
print('Logistic Regression Accuracy: ', accuracy_score(predictLR, y_test))

predictRF = modelRF.predict(x_test)
print('Random Forest Classifier : \n', classification_report(predictRF, y_test))
print('Random Forest Classifier Accuracy: ', accuracy_score(predictRF, y_test))

predictGB = modelGB.predict(x_test)
print('Gradient Boost Classifier : \n', classification_report(predictGB, y_test))
print('Gradient Boost Classifier Accuracy: ', accuracy_score(predictGB, y_test))
```

Recall
A Churn class Recall of 0.08 means that the model was able to catch 8% of the actual Churn cases. 
This is the measure we really care about, because we want to miss as few of the true Churn cases as possible.

Precision
Precision of the Churn class measures how often the model catches an actual Churn case, 
while also factoring in how often it misclassifies a Non-Churn case as a Churn case. 
In this case, a Churn Precision of 0.61 is not a problem because there are no significant consequences of identifying a customer as a Churn risk when she isn’t.

## We see that post the implementation of weights in the logistics regression classifier our accuracy reduces from initial 94% to 61% but our precision and recall both have improved for the minority class that is the churn user.
